```python
#import statements
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib nbagg
#%matplotlib inline
```

```python
data_file1 = pd.read_csv('datafile1.csv', header=None, sep=",")
data_file2 = pd.read_csv('datafile2.csv', header=None, sep=",")
data_file3 = pd.read_csv('datafile3.csv', header=None, sep=",")
data_file4 = pd.read_csv('datafile4.csv', header=None, sep=",")
data_file1.columns = ['x1', 'x2','y']
data_file2.columns = ['x1', 'x2','y']
data_file3.columns = ['x1', 'x2','y']
data_file4.columns = ['x1', 'x2','y']

```

```python
plt.scatter(data_file4['x1'], data_file1['x2'])
plt.show()
```

```python
def predict(x,s):
    activation = np.dot(x,s.T)
    if(activation > 0):
        return 1
    else:
        return 0
    
```

```python
def train(x,y):
    weights = np.random.random_sample([3])
    weights = np.mat(weights)
    epochs = 20
    l_rate = 0.01
    for x_row,y_row in zip(x,y):
        s = weights[-1]
        for i in range(epochs):              
            activation = predict(x_row,s)
            error = y_row[0]-activation
            s[0,0] = s[0,0] + (l_rate)*(error)*(x_row[0])
            s[0,1] = s[0,1] + (l_rate)*(error)*(x_row[1])
            s[0,2] = s[0,2] + (l_rate)*(error)          
            print('epoch = ',i,' w1 = ', weights[-1].item((0, 0)),' w2 = ', weights[-1].item((0, 1)),
                ' bias = ', weights[-1].item((0, 2)),' error = ',error)
            if(error == 0):
                break  
        print("\n")        
        weights = np.vstack([weights,s])
    weights = np.delete( weights , (0), axis=0)    
    return weights        
```

```python
def perceptron(data_file):
    data = data_file.copy()
    y_df  = pd.DataFrame(data_file['y'])
    data.drop(['y'],axis = 1,inplace = True)
    b = pd.Series(np.ones(data['x1'].size))
    data['b'] = b
    x = data.as_matrix()
    y = y_df.as_matrix()
    weights = train(x,y)
    return(weights)
```

```python
w_matrix1 = perceptron(data_file1)
```

```python
w_matrix2 = perceptron(data_file2)
```

```python
w_matrix3 = perceptron(data_file3)
```

```python
w_matrix4 = perceptron(data_file4)
```

```python
#animated plot for data_file1
fig = plt.figure()
ax = plt.axes(xlim=(-2, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

ax.scatter(data_file1['x1'], data_file1['x2'], c = data_file1['y'])
# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = data_file1['x1']
    y = -(w_matrix1[i,2]/w_matrix1[i,1]) -((w_matrix1[i,0]*x)/w_matrix1[i,1])
    line.set_data(x, y)
    return line,


# call the animator.  blit=True means only re-draw the parts that have changed.
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=len(w_matrix1), interval=100, blit=True,repeat = False)
plt.show()
```
